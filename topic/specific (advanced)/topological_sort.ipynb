```python
## Course Schedule

# trying to detect a cycle
# first convert into adjacency graph structure

# recursive dfs is easy to write
# add a path boolean array, refresh after calling finishing recursive stack resolves

# def recursive_dfs(v, visited, graph):
#     if not visited[v]:
#         visited[v] = True
#         for w in graph[v]:
#             recursive_dfs(w, visited, graph)

class Solution:
    def __init__(self) -> None:
        self.cycle = False

    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        self.path = [False] * numCourses
        self.visited = [False] * numCourses
        
        graph = self._graph(numCourses, prerequisites)

        for i in range(numCourses):
            self._dfs(graph, i)
        
        return not self.cycle

    def _graph(self, numCourses, prerequisites):
        graph = [[] for _ in range(numCourses)]
        for (after, before) in prerequisites:
            graph[before].append(after)
        return graph

    def _dfs(self, graph, start):
        if self.path[start]:
            self.cycle = True

        if self.visited[start] or self.cycle:
            return

        self.visited[start] = True

        self.path[start] = True
        for child in graph[start]:
            self._dfs(graph, child)
        self.path[start] = False
            
    
```

```python
## Course Schedule 2

# Topological sort DFS
# Find a linear ordering of the directionality of the nodes (cannot be done if there is a cycle)
# Simply reverse the postorder traversal (post order means left -> right -> node)
# In post order children are before parents
# Topological sort requires parents before children

# DFS preorder -> before children loop
# DFS postorder -> after children loop

class Solution:
    def __init__(self) -> None:
        self.cycle = False
        self.postorder = []
    
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        self.path = [False] * numCourses
        self.visited = [False] * numCourses
        
        graph = self._graph(numCourses, prerequisites)

        for i in range(numCourses):
            self._dfs(graph, i)
        
        if self.cycle:
            return []

        return reversed(self.postorder)

    def _graph(self, numCourses, prerequisites):
        graph = [[] for _ in range(numCourses)]
        for (after, before) in prerequisites:
            graph[before].append(after)
        return graph

    def _dfs(self, graph, start):
        if self.path[start]:
            self.cycle = True

        if self.cycle or self.visited[start]:
            return

        self.visited[start] = True
        self.path[start] = True

        for child in graph[start]:
            self._dfs(graph, child)

        self.postorder.append(start)
        self.path[start] = False
```

```python
# Kahns Algorithm
# BFS topological sort

# count all indegree with the edges
# add 0 to the queue
# while q:
# pop and decrement ingoing edges
# add 0 to the q
# record sequence of cur nodes

class Kahns:
    
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        graph = self._graph(numCourses, prerequisites)
        indegree = [0] * numCourses

        for req, edge  in prerequisites:
            indegree[edge] += 1
        
        queue = deque()

        for i in range(numCourses):
            if indegree[i] == 0:
                queue.append(i)
        
        res = [0] * numCourses
        count = 0

        while queue:
            cur = queue.popleft()
            ## this is the topological sort sequence
            res[count] = cur
            count += 1
            for next in graph[cur]:
                # we delete the node, and remove all outgoing edges from the node
                indegree[next] -= 1
                if indegree[next] == 0:
                    queue.append(next)
        
        # there is a cycle
        if count != numCourses:
            return []
        
        return res

    
    def _graph(self, numCourses, prerequisites):
            graph = [[] for _ in range(numCourses)]
            for (after, before) in prerequisites:
                graph[before].append(after)
            return graph
        
```
