```python
# max area of island 
# DFS

def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
    x, y = len(grid), len(grid[0])
    visit = set()

    def search(i, j):
        if (
            i < 0 or 
            j < 0 or 
            i == x or 
            j == y or
            grid[i][j] == 0 or
            (i, j) in visit
        ): 
            return 0
        else:
            visit.add((i, j))
            return 1 + search(i+1, j) + search(i, j+1) + search(i-1, j) + search(i ,j-1)

    area = 0

    for i in range(x):
        for j in range(y):
            area = max(area, search(i, j))

    return area
```

```python
# rotting oranges
# BFS

def orangesRotting(self, grid: List[List[int]]) -> int:
    time = 0
    fresh = 0
    q = deque()
    
    max_x, max_y = len(grid), len(grid[0])

    for i in range(max_x):
        for j in range(max_y):
            if grid[i][j] == 1:
                fresh += 1
            if grid[i][j] == 2:
                q.append((i, j))

    while q and fresh > 0:
        for _ in range(len(q)):
            i, j = q.popleft()

            for x,y in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
                new_x = i+x
                new_y = j+y

                if (
                    new_x in range(max_x) and
                    new_y in range(max_y) and
                    grid[new_x][new_y] == 1
                ):
                    grid[new_x][new_y] = 2
                    q.append((new_x, new_y))
                    fresh -= 1
        
        time += 1
    
    return time if fresh == 0 else -1
```

```python
# 01 mat
# just use BFS, since each tile is deeply connected to each other
# this is classic bfs problem

def dist(matrix):
    n, m = len(matrix), len(matrix[0])
    
    res = [[-1 for _ in range(m)] for _ in range(n)]

    q = deque()

    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                q.append((i, j))
                res[i][j] = 0
    
    dirs = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    while q:
        x, y = q.popleft()
        for dx, dy in dirs:
            next_x, next_y = x + dx, y + dy
            if 0 <= next_x < n and 0 <= next_y < m and res[next_x][next_y] == -1:
                res[next_x][next_y] = res[x][y] + 1
                q.append((next_x, next_y))
    
    return res
```

```python
## Number of Islands

# loop through every cell
# have a visited
# do DFS everytime you see a 1

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        res = 0
        
        n = len(grid)
        m = len(grid[0])
        
        for i in range(n):
            for j in range(m):
                if grid[i][j] == '1':
                    res += 1
                    self.dfs(grid, i, j, n, m)
        
        return res
    
    def dfs(self, grid, x, y, n, m):
        if x >= n or x < 0 or y >= m or y < 0:
            return
        
        if grid[x][y] == '0':
            return
    
        grid[x][y] = '0'
    
        self.dfs(grid, x+1, y, n, m)
        self.dfs(grid, x, y+1, n, m)
        self.dfs(grid, x-1, y, n, m)
        self.dfs(grid, x, y-1, n, m)
```

```python
# word search

# just do dfs once you match
# use '#' marker for temporary marker for visited
# revert after dfs loop

class Solution:
    def __init__(self):
        self.found = False

    def exist(self, board: List[List[str]], word: str) -> bool:
        m, n = len(board), len(board[0])

        for i in range(m):
            for j in range(n):
                self.dfs(board, i, j, word, 0)

                if self.found:
                    return True
        
        return False
    
    def dfs(self, board, i, j, word, match):
        if self.found:
            return

        if i < 0 or i >= len(board) or j < 0 or j >= len(board[0]):
            return
        
        if board[i][j] != word[match]:
            return
        
        if match == len(word) - 1:
            self.found = True
            return
        
        temp = board[i][j]
        board[i][j] = '#'

        self.dfs(board, i+1, j, word, match+1)
        self.dfs(board, i, j+1, word, match+1)
        self.dfs(board, i-1, j, word, match+1)
        self.dfs(board, i, j-1, word, match+1)

        board[i][j] = temp
```
