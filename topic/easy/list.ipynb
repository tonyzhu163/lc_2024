```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# slow and fast pointer
# walk fast to n ahead of slow
# walk both to the end
# remove slow

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        lst = ListNode(-1)
        lst.next = head
        
        slow = self.walk(lst, n+1)
            
        slow.next = slow.next.next
        
        return lst.next
    
    def walk(self, head, n):
        slow = head
        fast = head
        
        while n > 0:
            fast = fast.next
            n -= 1
        
        while fast != None:
            slow = slow.next
            fast = fast.next
        
        return slow
        
```

```python
# sort a linkedlist

# mergesort

class Solution:
    def sort(self, head):
        left = head
        right = self.getmid(head)
        tmp = right.next
        right.next = None
        right = tmp

        # we need to split the lists at the middle node
        # left = head -> xxx -> right -> None
        # right = right -> xxx -> None

        left = self.sort(left)
        right = self.sort(right)
        return self.merge(left, right)

    def getmid(self, head):
        slow, fast = head, head.next

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        return slow

    def merge(self, l1, l2):
        dummy = cur = ListNode()
        
        while l1 and l1:
            if l1.val < l2.val:
                cur.next = l1
                l1 = l1.next
            else:
                cur.next = l2
                l2 = l2.next
        
        if l1:
            cur.next = l1
        if l2:
            cur.next = l2
        
        return dummy.next
```

```python
# merge k sorted lists

# just use a priority queue

# just push all the heads
# keep popping from PQ into node, push node.next
# and move it forward while PQ is nonempty

from queue import PriorityQueue

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None

        dummy = cur = ListNode()
        
        pq = PriorityQueue()
        
        # tie breaker
        counter = 0

        for head in lists:
            if head is not None:
                pq.put((head.val, counter, head))
                counter += 1
        
        while not pq.empty():
            _, _, next_node = pq.get()
            cur.next = next_node
            if next_node.next:
                pq.put((next_node.next.val, counter, next_node.next))
                counter += 1
            cur = cur.next
        
        return dummy.next
        
        
```
